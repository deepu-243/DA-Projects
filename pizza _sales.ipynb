# Total sales

```python
df= pd.merge(left= p, right=od, on = "pizza_id", how = "inner")
df
```

```python
df["Amounts"]= df["price"]*df["quantity"]
df
```

```python
c=df["Amounts"].sum()
print("OUR TOTAL SALES IS ", c)
```

# Most ordered pizza

```python
df2 = pd.merge(left = df, right =pt, on= "pizza_type_id", how = "inner")
df2
```

```python
gb = df2.groupby("name").agg({"order_details_id":"count"})
gb
```

```python
m = gb["order_details_id"].idxmax()
print("Mostly Ordered Pizza Is", m)
```

# Identify the time at which we are getting high orders and the time at which we are getting the lowest orders

```python
ord
```

```python
ord["hour"]= ord["time"].dt.strftime("%H")
ord
```

```python
gb4= ord.groupby("hour").agg({"order_id":"count"})
gb4
```

```python
sns.lineplot(x = gb4.index, y = gb4["order_id"], marker = "o")
plt.show()
```

```python
max= gb4["order_id"].idxmax()
min=gb4["order_id"].idxmin()

print("Bussiest hours: ", max)
print("Idle hours: ", min)
```

# Identify at which time of the day, people are ordering more pizzas. (morning/evening/afternoon/night)

```python
ord["hour"]= ord["time"].dt.strftime("%H")
ord
```

```python
ord["hour"]= ord["hour"].astype(int)
ord.info()
```

```python
ord.loc[(ord["hour"] < 12),"Shift"] = "morning"
ord.loc[((ord["hour"] >=12)&(ord["hour"]<16),"Shift")]= "Afternoon"
ord.loc[((ord["hour"] >=16)&(ord["hour"]<21),"Shift")] = "Evening"
ord.loc[((ord["hour"] >=21),"Shift")]="Night"
```

```python
ord
```

```python
sns.countplot(x = "Shift", data = ord)
```
