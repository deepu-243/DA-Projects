```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading data
USA_data = pd.read_excel('Superstore_USA.xlsx')
```

```python
USA_data.head()
```

```python
#Checked Whether there are null values or not
USA_data.isnull().sum()
```

**There are 72 null values in Product Base Margin Column**

```python
#Checking datatypes of each column
USA_data.info()
```

```python
#Replacing null values with 0
USA_data['Product Base Margin'].fillna(0)
```

```python
#Checking unique values in Product Category
USA_data['Product Category'].unique()
```

```python
#Checking Whether there are any null values left or not
USA_data['Product Base Margin'].isnull().sum()
```

```python
USA_data.describe()
```

```python
#Checking unique values in Product Sub-Category
USA_data['Product Sub-Category'].unique()
```

```python
#Checking total Order Count
USA_data['Order ID'].count()
```

```python
#Checking Orders per city
orders_per_city = USA_data.groupby(['City']).agg({'Order ID': 'count'})[:10]
orders_per_city
```

```python
#Plotting Bar Chart of orders_
orders_per_city.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Orders')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Checking Orders_per_region
orders_per_region = USA_data.groupby(['Region']).agg({'Order ID': 'count'})
orders_per_region
```

```python
#Plotting Bar Chart of orders_
orders_per_region.plot(kind='bar',color = 'darkblue')
plt.xlabel('Region')
plt.ylabel('Orders')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Lets see how sub-categories are distributed wrt to category
plt.figure(figsize=(16,8))
plt.bar('Product Sub-Category','Product Category',data=USA_data,color='yellow')
plt.xticks(rotation = 90)
plt.show()
```

```python
USA_data['Product Sub-Category'].unique()
```

```python
orders_sub_category = USA_data.groupby(['Product Sub-Category']).agg({"Order ID": "count"})
orders_sub_category
```

```python
orders_sub_category.plot(kind='bar')
plt.ylabel('Orders')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.pie(orders_sub_category['Order ID'], labels=orders_sub_category.index,autopct='%1.0f%%')
plt.show()
```

**The store has wide variety of Office Supplies especially in Binders and Paper department.**

```python
#Total Profit and Sales per Sub-Category
a=USA_data.groupby('Product Sub-Category')[['Sales', 'Profit']].agg('sum')
a
```

```python
#Total Profit and Sales per Sub-Category
a.plot(kind='bar')
plt.title('Total Profit and Sales per Sub-Category')
plt.legend(['Sales', 'Profit'])
plt.show()
```

```python
#Distribution of Top 10 Products
plt.figure(figsize=(12,10))
USA_data['Product Name'].value_counts().head(10).plot.pie(autopct="%1.1f%%")
plt.ylabel('Count')
plt.show()
```

```python
#Count of Sub-Category region wise
plt.figure(figsize=(15,8))
sns.countplot(x="Product Sub-Category", hue="Region", data=USA_data)
plt.xticks(rotation=90)
plt.show()
```

**People residing in Central part of US tend to order more from superstore.**

```python
#extracting the year of order
USA_data['order year']=USA_data['Order Date'].dt.year
USA_data['order year'].head()
```

```python
#Sales per year
b= USA_data.groupby('order year')[['Sales','Profit']].agg(['sum'])
b
```

```python
b.plot(kind='bar')
plt.title('Year wise Total Sales & % of profit gained')
plt.show()
```

```python
#Top 10 customers who order frequently
top10_cust = USA_data['Customer Name'].value_counts().head(10)
top10_cust
```

```python
top10_cust.plot(kind='bar')
plt.ylabel('Orders')
plt.show()
```

```python
USA_data['Order Priority'] = USA_data['Order Priority'].replace("Critical ", "Critical")
```

```python
#Orders with respect to their priority
order_priority = USA_data.groupby('Order Priority').agg({"Order ID" : 'count'})
order_priority
```

```python
plt.pie(x=order_priority['Order ID'], labels =order_priority.index,autopct='%1.0f%%' )
plt.show()
```

```python
#orders w.r.t ship mode
orders_mode = USA_data.groupby("Ship Mode").agg({'Order ID': 'count'})
orders_mode
```

```python
plt.pie(x = orders_mode['Order ID'], labels = orders_mode.index, autopct = '%1.0f%%')
plt.show()
```

```python
#Orders w.r.t customer segment
orders_segment = USA_data.groupby('Customer Segment').agg({'Order ID': 'count'})
orders_segment
```

```python
plt.pie(x = orders_segment['Order ID'], labels = orders_segment.index, autopct = '%1.0f%%')
plt.show()
```
